## Function

```python
# function is block of statements defined for a specific task.
# advantages--> 1. Modularity, 2. Readability, 3. Reusability, 4. Reduction is code size, 5. Abstraction 
# syntax
# def  function_name(listof arguments):
#    statements
#    statements
#    return values/variables
```

```python
def circle(rad):
    pi = 3.14
    area = pi * rad * rad
    perimeter = 2 * pi * rad
    return area , perimeter
```

```python
circle(3)
```

```python
x, y  = circle(3)  # function calling
print("Area of circle is:", x)
print("Perimeter of circle is:", y)
```

### insertion

```python
lst = [10, 20, 30, 40, 50, 60]
print(lst)

# append(value) --->insert at the end of list
lst.append(70)
print(lst)

# insert(pos, value)---> insert a value at pos
lst.insert(1, 80)
print(lst)

# extend(list)---> add the elements of one list to another elementwise
l = [100, 200, 300]
lst.extend(l)
print(lst)
```

```python
l = [100, 200, 300]
lst = [10, 20, 30, 40, 50, 60]
pos = 1
for i in l:
    lst.insert(pos, i)
    pos = pos + 1
print(lst)
```

### deletion

```python
lst = [10, 20, 30, 40, 30, 50, 60]
print(lst)

# pop() --> deletes the last element from the list
lst.pop()
print(lst)

# pop(pos)--> deletes a specific element from the pos.

lst.pop(1)
print(lst)

# remove(value)--> deletes the first occurence of the element in the list
lst.remove(30)
print(lst)

# clear()--> to delete all the elements
lst.clear()
print(lst)
```

```python
# count() ---> frequency of the element
lst = [10, 20,30, 40, 30, 50, 30, 20, 30, 80]
print(lst)
c = lst.count(30)
print(c)
for i in range(c):
    lst.remove(30)

print(lst)
```

```python
# count() ---> frequency of the element
lst = [10, 20,30, 40, 30, 50, 30, 20, 30, 80]
print(lst)
c = lst.count(30)
print(c)
while(c>0):
    lst.remove(30)
    c = c-1
print(lst)
```

```python
lst = [10, 20,30, 40, 30, 50, 30, 20, 30, 80]
print(lst)
for i in lst:
    if i == 30:
        lst.remove(30)
print(lst)
```
